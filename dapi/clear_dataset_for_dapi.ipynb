```python
import os

# Ignore text contains all bad probes
file_path = r'E:\Data_sets\ISL\ignore.txt' 

# File path to the dataset folder
folder_path = r'E:\Data_sets\ISL\tiff'

# Read all bad probes and all dataset files
files = os.listdir(folder_path)
with open(file_path, 'r') as file:
    lines = file.readlines()  # Read all lines into a list
```

```python
# Remove bad images which have poor quality, incorrect microscope settings ...
remove_count = 0
for line in lines:
    search_string = line.strip()
    for file in files:
        if search_string in file:
            file_path = os.path.join(folder_path, file)
            if os.path.exists(file_path):
                os.remove(file_path)
                remove_count += 1     
                print(f"Removed {file_path}")
print(f"Removed {remove_count} files")
```

```python
# ch1 = Contrast ; ch2 = Cy3 ; ch3 = GFP ; ch4 = DAPI ; ch5 = BrightField
# Removes all ch2, ch3 data since we don't need them for training
search_strings = ["ch2.tiff","ch3.tiff"]
remove_count = 0
for file in files:
    if any(search_string in file for search_string in search_strings):
        # Construct the full path to the file
        file_path = os.path.join(folder_path, file)
        # Remove the file
        if os.path.exists(file_path):
            os.remove(file_path)
            remove_count+=1
            print(f"Removed {file_path}")
print(f"Removed {remove_count} files")
```

```python
# Create DAPI dataset
important_data = []
dapi_tiff_path = r"E:\Data_sets\ISL\tiff_dapi"
main_path = r"E:\Data_sets\ISL\split_dapi"
path_suffix = ["train.txt","test.txt","val.txt"]
```

```python
all_probes = []
for suffix in path_suffix:
    path = os.path.join(main_path, suffix)
    print(path)
    with open(path, 'r') as file:
        lines = file.readlines()  # Read all lines into a list
    
    for line in lines:
        print(line.strip())
        all_probes.append(line.strip())
```

```python
import shutil
source_folder = r'E:\Data_sets\ISL\tiff'
destination_folder = dapi_tiff_path

# Ensure the destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Initialize a list to keep track of copied files for reporting
copied_files = []

# Loop over each file in the source folder
for file in os.listdir(source_folder):
    # Check if the file name contains any of the probes
    if any(element in file for element in all_probes):
        # Construct the full path to the source and destination files
        source_path = os.path.join(source_folder, file)
        destination_path = os.path.join(destination_folder, file)
        
        # Copy the file
        shutil.copy2(source_path, destination_path)
        copied_files.append(file)
```
